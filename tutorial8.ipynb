## Functions in Python 

### Calling a Function

```python
def myfun(x, y):
    return x * y
```

```python
myfun(3, 4)
```

### Functions are first-class objects in Python

```python
def myfun(x):
    return x*3
```

```python
def apply(q, x):
    return q(x)
```

```python
apply(myfun, 7)
```

### Lambda Notation

```python
# 간단한 버전
apply(lambda z: z * 4, 7)
```

### Default Values for Arguments

```python
def myfun(b, c=3, d="hello"):
    return b + c
```

```python
myfun(5,3,"hello")
```

```python
myfun(5,3)
```

```python
myfun(5)
```

### Keyword Arguments

```python
def myfun (a, b, c):
    return a-b
```

```python
myfun(2, 1, 43)
```

```python
myfun(c=43, b=1, a=2)
```

```python
myfun(2, c=43, b=1)
```
