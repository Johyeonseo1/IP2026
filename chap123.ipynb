## **(chap1.2.3)** Drawing Functions in OpenCV

```python
import numpy as np
import cv2
```

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)
```

선을 그리기

```python
size = 5 # 선의 사이즈
Blue, Green, Red = 255, 0, 0 # 선의 색상
start_x, start_y = 0, 0 # 선의 시작점
end_x, end_y = 511, 511 # 선의 마지막점
img = cv2.line(img, (start_x, start_y), (end_x, end_y), (Blue, Green, Red), size)
```

사각형 그리기

```python
img = cv2.rectangle(img,(384,0),(510,128),(0,255,0),3)
```

원을 그리기

```python
center_x, center_y = 447, 63 # 원의 중심점
radius = 63 # 원 반지름
thickness = -1 # 원의 뚜께 -1이면 다 채움
img = cv2.circle(img, (center_x, center_y), radius, (0,0,255), thickness)
```

타원 그리기

```python
x_axis, y_axis = 100, 50 # 타원의 크기
angle = 0 # 기울기
start_angle = 0 # 시작 각도
end_angle = 180 # 시작 각도에서 부터 해당 각도까지 그려줌
img = cv2.ellipse(img,(256,256), (x_axis, y_axis), angle, start_angle, end_angle, 255, -1)
# 색상은 1개만 넣으면 파랑색으로 나옴
```

다각형 그리기

```python
closed = True # 점들이 연결되어서 선으로 나올지 여부
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
img = cv2.polylines(img,[pts],closed,(0,255,255))
```

텍스트를 그리기

```python
font = cv2.FONT_HERSHEY_SIMPLEX # 글골
font_scale = 4 # 기본 크기의 배수
cv2.putText(img,'OpenCV',(10,500), font, font_scale,(255,255,255),2,cv2.LINE_AA)
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
