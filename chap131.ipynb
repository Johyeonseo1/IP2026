## Basic Operations on Images

```python
import cv2
import numpy as np

img = cv2.imread('images.jpg')
```

```python
px = img[100,100]
print ("해당 좌표의 BGR색상 값 :",px)

blue = img[100,100,0]
print ("파랑색 값 :",blue)
```

```python
# 해당 좌표 하양색으로 수정
img[100:200,100:200] = [255,255,255]
# img[10:500,10:500, 2] = [255] # 해당 영역을 빨강색으로
print (img[100,100])
```

```python
print(img.item(10,10,2))

img.itemset((10,10,2),100)
print(img.item(10,10,2))
```

```python
print (img.shape)
```

```python
print (img.size)
```

```python
print (img.dtype)
```

### Image ROI

```python
ball = img[280:340, 330:390]
img[273:333, 100:160] = ball
```

### Splitting and Merging Image Channels

```python
b,g,r = cv2.split(img)
img = cv2.merge((b,g,r))
```

```python
b = img[:,:,0]
```

```python
img[:,:,2] = 0
```
