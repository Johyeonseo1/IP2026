## **(chap1.2.1)** Read an image

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# 이미지를 불러옴
img = cv2.imread('images.jpg', 0) # 0으로하면 흑백 
```

### Display an image

```python
name = 'image' # 이미지 창의 이름
cv2.imshow(name, img)
cv2.waitKey(0) #키 입력 대기중
cv2.destroyAllWindows() #이미지 창을 닫음
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Write an image

```python
# save_name의 이름으로 이미지를 저장함
save_name = 'messigray.png'
cv2.imwrite(save_name, img)
```

### Sumit up

```python
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27: # ESC키를 누르면 종료
    cv2.destroyAllWindows()
elif k == ord('s'): # 's'를 누르면 'messigray1'이름으로 저장
    cv2.imwrite('messigray1.png',img)
    cv2.destroyAllWindows()
```

### Using Matplotlib

```python
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
plt.show()
```
