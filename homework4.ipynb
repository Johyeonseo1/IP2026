```python
import cv2
import numpy as np

font = cv2.FONT_HERSHEY_SIMPLEX

def run_main():
    cap = cv2.VideoCapture("Coin.mp4")
    size = (700, 500)

    while(True):
        ret, frame = cap.read()
        if ret == False:
            break

        count = 0 
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
        gray_blur = cv2.GaussianBlur(gray, (5, 5), 2)
        thresh = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV, 21, 5)

        kernel = np.ones((5, 5), np.uint8)
        closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE,kernel, iterations=2)

        cont_img = closing.copy()
        contours, hierarchy = cv2.findContours(cont_img, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

        for cnt in contours:
            area = cv2.contourArea(cnt)
            if area < 16000 or area > 26000:
                continue
            if len(cnt) < 5:
                continue
            count += 1
            ellipse = cv2.fitEllipse(cnt)
            cv2.ellipse(frame, ellipse, (0,255,0), 2)

        cv2.putText(frame,str(count) ,(20,50), font, 2,(0,255,0),3,cv2.LINE_AA)

        resized_closing = cv2.resize(closing, size)
        resized_thresh = cv2.resize(thresh, size)
        resized_frame = cv2.resize(frame, size)

        cv2.imshow("Morphological Closing", resized_closing)
        cv2.imshow("Adaptive Thresholding", resized_thresh)
        cv2.imshow('Contours', resized_frame)

        if cv2.waitKey(15) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    run_main()
```
