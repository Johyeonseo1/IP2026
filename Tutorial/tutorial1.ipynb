```python
t = (23, 'abc', 4.56, (2,3), 'def')
print('t[1] :', t[1])
print('t[-3] :', t[-3])
```

```python
t = (23, 'abc', 4.56, (2,3), 'def')
print('t[1:4] :', t[1:4])
print('t[1:-1] :', t[1:-1])
print('t[1:-1:2] :', t[1:-1:2])
```

```python
t = (23, 'abc', 4.56, (2,3), 'def')
print('t[:2] :', t[:2])
print('t[2:] :', t[2:])
print('t[:] :', t[:])
```

```python
list1 = [1, 2, 3]
list2 = list1
print(list2)
list2 = list1[:]
print(list2)
```

```python
t = [1, 2, 4, 5]
print('3 in t :', 3 in t)
print('4 in t :', 4 in t)
print('4 not in t :', 4 not in t)
```

```python
a = 'abcde'
print("'c' in a :", 'c' in a)
print("'cd' in a :", 'cd' in a)
print("'ac' in a :", 'ac' in a)
```
