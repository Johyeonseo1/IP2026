```python
import cv2
import numpy as np
```

```python
images = 'images.jpg'
font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.imread(images)
drawing = False
mode = True
ix,iy = 0, 0
lx,ly = 0, 0
rgbg = 2
```

```python
def nothing(x):
    pass

def draw_circle(event,x,y,flags,param):
    global ix,iy,lx,ly,drawing,mode,img

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            lx,ly = x,y
            cv2.rectangle(img,(ix,iy),(lx,ly),(255,255,255),-1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False

```

```python
cv2.namedWindow('image')
cv2.createTrackbar('value','image',0,255,nothing)
cv2.setMouseCallback('image',draw_circle)

while(1):
    cv2.imshow('image',img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    elif k == ord('b'): 
        rgbg = 0
    elif k == ord('g'): 
        rgbg = 1
    elif k == ord('r'): 
        rgbg = 2
    

    color = cv2.getTrackbarPos('value','image')
    img = cv2.imread(images)
    
    img[iy:ly, ix:lx, rgbg] = [255 - color]
    test = 'Mouse position ('+str(ix)+','+str(ix)+') ('+str(lx)+','+str(lx)+') - '+str(color)+ ' - '+ str(rgbg)
    cv2.putText(img,test,(10,30), font, 0.7,(255,255,255),1,cv2.LINE_AA)
cv2.destroyAllWindows()
```
