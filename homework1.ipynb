```python
import numpy as np
import cv2
```

### 변수 선언

```python
img = cv2.imread('image1.jpg')
font = cv2.FONT_HERSHEY_SIMPLEX # 글골
drawing = False
img_copy = img.copy()
begin_x, begin_y = 0, 0 # 사각형 시작 좌표
end_x, end_y = 0, 0 # 사각형 끝 좌표
color_info = 'White' # 컬러 변경할 색상명
color = (255,255,255) # 컬러 색상 정보
```

### 선택한 영역을 투명도가 적용된 사각형을 추가하는 함수

```python
def alpha_rectangle(begin_y,y,begin_x,x):
    global img, alpha
    # 좌표 정렬
    y1, y2 = min(begin_y, y), max(begin_y, y) 
    x1, x2 = min(begin_x, x), max(begin_x, x)
    
    if y2 > y1 and x2 > x1:
        img = img_copy.copy()
        rectangle_img = img[y1:y2, x1:x2]
        rectangle_color = np.full(rectangle_img.shape, color, dtype=np.uint8)
        result = cv2.addWeighted(rectangle_color, alpha, rectangle_img, 1.0 - alpha, 0)
        img[y1:y2, x1:x2] = result
    return img
```

### 마우스 상태에 따라 사각형을 그리는 함수

```python
def draw_circle(event,x,y,flags, param):
    global begin_x, begin_y,end_x, end_y,drawing,img

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        img = img_copy.copy()
        # 초기화
        begin_x, begin_y = x, y
        end_x, end_y = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            end_x, end_y = x, y

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
```

### 색상을 변경 시 사용하는 함수

```python
def change_color(key,info,a,b,c):
    global color, color_info, k
    if k == ord(key):
        color_info = info
        color = (a,b,c)
```

### 메인

```python
def nothing(x):
    pass

cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.createTrackbar('value','image',0,100,nothing) # 트랙바 추가
cv2.setMouseCallback('image',draw_circle)
cv2.setTrackbarPos('value', 'image', 50)

while(1):
    val = cv2.getTrackbarPos('value','image')
    alpha = val / 100.0
    test = 'Mouse position ('+str(begin_x)+','+str(begin_y)+') ('+str(end_y)+','+str(end_x)+') - '+str(alpha)+' - '+str(color_info)
    cv2.putText(img,test ,(10,30), font, 1,(255,255,255),1,cv2.LINE_AA)
    cv2.imshow('image',img)

    alpha_rectangle(begin_y,end_y,begin_x,end_x)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    change_color('k', 'Black', 0, 0, 0)
    change_color('b', 'Blue', 255, 0, 0)
    change_color('g', 'Green', 0, 255, 0)
    change_color('r', 'Red', 0, 0, 255)
    change_color('w', 'White', 255, 255, 255)

cv2.destroyAllWindows()
```
