```python
import cv2
import numpy as np
```

```python
color_info = 2 # 0 = B, 1 = G, 2 = R
font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.imread('image_1.jpg')
drawing = False
start_x, start_y = -1, -1
end_x, end_y = -1, -1
```

```python
def nothing(x):
    pass

def draw_circle(event,x,y,flags,param):
    global start_x, start_y,end_x, end_y,drawing,img
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        start_x, start_y = x, y
        end_x, end_y = -1, -1
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            end_x, end_y = x,y
        if end_x == -1 and end_y == -1:
            start_x, start_y = x, y
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
```

```python
cv2.namedWindow('image')
cv2.createTrackbar('value','image',0,255,nothing)
cv2.setMouseCallback('image',draw_circle)

while(1):
    cv2.imshow('image',img)

    color = cv2.getTrackbarPos('value','image')
    img = cv2.imread('image_1.jpg')
    
    y1, y2 = min(start_y, end_y), max(start_y, end_y) 
    x1, x2 = min(start_x, end_x), max(start_x, end_x)
    img[y1:y2, x1:x2, color_info] = [255 - color]

    text = 'Mouse position ('+str(max(0, start_x))+', '+str(max(0, start_y))+') ('+str(max(0, end_x))+', '+str(max(0, end_y))+') - '+str(color)+ ' - '+ str(color_info)
    cv2.putText(img,text,(10,30), font, 0.7,(255,255,255),1,cv2.LINE_AA)

    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
```
