```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('Dice.mp4')

while(1):
    # Take each frame
    a, frame = cap.read()
    if a == False:
        break

    # Convert BGR to HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    # define range of blue color in HSV
    lower_blue = np.array([100,50,10])
    upper_blue = np.array([150,255,255])
    # Threshold the HSV image to get only blue colors
    mask = cv2.inRange(hsv, lower_blue, upper_blue)
    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(frame,frame, mask= mask)

    # 각 마스크 크기 조절
    resized_frame = cv2.resize(frame, (700, 500))
    resized_mask = cv2.resize(mask, (700, 500))
    resized_res = cv2.resize(res, (700, 500))

    cv2.imshow('frame',resized_frame)
    cv2.imshow('mask',resized_mask)
    cv2.imshow('res',resized_res)

    k = cv2.waitKey(15) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
```
